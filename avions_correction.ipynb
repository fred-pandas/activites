```python
from random import random
```

```python
def attente():
    y = 0
    while random() > p:
        y = y+1
    return y
```

```python
attente()
```

```python
def echan(n):
    return [attente() for k in range(n)]
```

```python
def plotbis(n):
    pl.hist(Delta,bins=[k-0.5 for k in range(50)],
            edgecolor='black')
    pl.hist(echan(n),bins=[k-0.5 for k in range(50)],
            alpha=0.5,color='red',edgecolor='black')
```
